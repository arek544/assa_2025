# Setup environment

```python
# Cell to install required packages and verify imports
import sys
import subprocess

packages = [
    'numpy',
    'librosa',
    'seaborn',
    'matplotlib',
    'scikit-learn',
    'ipython',
    'keras',
    'tensorflow',
    'soundfile'
]

# Install any missing packages
for pkg in packages:
    try:
        __import__(pkg if pkg != 'scikit-learn' else 'sklearn')
        print(f"{pkg}: already installed")
    except Exception:
        print(f"Installing {pkg}...")
        subprocess.check_call([sys.executable, '-m', 'pip', 'install', pkg])

# Verify imports
for pkg in packages:
    module_name = pkg if pkg != 'scikit-learn' else 'sklearn'
    try:
        __import__(module_name)
        print(f"SUCCESS: imported {module_name}")
    except Exception as e:
        print(f"FAIL: could not import {module_name}: {e}")
```
